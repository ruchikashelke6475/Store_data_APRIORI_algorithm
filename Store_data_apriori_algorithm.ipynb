```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv('store_data.csv', header = None )
```

```python
file_path = 'store_data.csv'
```

```python
dataset= [] 
with open(file_path)as fp:
    for line in fp:
        dataset.append([i.strip() for i in line.strip().split(',')])
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
```

```python
print(te_ary)
```

```python
arr = np.array(te_ary, dtype=int)
df = pd.DataFrame(arr, columns =te.columns_)
```

## Association Rules

```python

def build_assiciation_rules(df,min_support = 0.05, min_threshold = 1, metric ='lift'):
    from mlxtend.frequent_patterns import apriori,association_rules
    frequent_itemset = apriori(df,min_support=min_support, use_colnames= True)
    rules = association_rules(frequent_itemset,metric=metric, min_threshold=min_threshold)
    return rules
    
```

```python
rules = build_assiciation_rules(df)
```
